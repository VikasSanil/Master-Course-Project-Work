```python
import warnings
warnings.filterwarnings("ignore")

# Logistic regression and SVM were originally designed for binary classifications
# They can also be used for multi-class classifications
# They require a numerical feature space to be run. Normalization is not required

import pandas as pd
from sklearn import preprocessing
from IPython.display import display, HTML

df=pd.read_csv('data_students_10k.csv')
print(df.shape)
# strip column names
df=df.rename(columns=lambda x: x.strip())
cols=df.columns
# print out and display dataframe as tables in HTML
display(HTML(df.head(10).to_html()))

# replace missing values in numerical variables by using mean value #################################
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Hours on Assignments"].fillna(df["Hours on Assignments"].mean(), inplace=True)
df["Hours on Games"].fillna(df["Hours on Games"].mean(), inplace=True)
df["Exam"].fillna(df["Exam"].mean(), inplace=True)
df["Grade"].fillna(df["Grade"].mean(), inplace=True)

# check again whether there are missing values
print('ColumnName, DataType, MissingValues')
for i in cols:
    print(i, ',', df[i].dtype,',',df[i].isnull().any())
 
# remove column ID and grade which are not appropriate to be included in this classification task
df=df.drop('ID',1)
df=df.drop('Grade',1)

# encode labels
y = df['GradeLetter'] # define label as nominal values
le = preprocessing.LabelEncoder()
le.fit(y)
y_encoded = le.transform(y) # encode nominal labels to integers #####################################
df['GradeLetter'] = y_encoded

# print out and display dataframe as tables in HTML
display(HTML(df.head(10).to_html()))

print('Column Datatypes:\n',df.dtypes)

# convert all nominal variables to binary variables
df_num=df.copy(deep=True) 
# create new binary columns
df_dummies=pd.get_dummies(df_num[['Degree','Nationality']])
# add them to dataframe
df_num=df_num.join(df_dummies)
# drop original columns
df_num=df_num.drop('Degree',axis=1)
df_num=df_num.drop('Nationality', axis=1)

# drop extra binary columns, since we only need N-1 binary columns
df_num=df_num.drop('Degree_ BS', axis=1)
df_num=df_num.drop('Nationality_ China', axis=1)

display('df_num:',HTML(df_num.head(10).to_html()))
```

```python
# 01. Solutions for imbalance issues

# First, check the distribution of labels
print(df_num.groupby(['GradeLetter']).size())

# Note that imbalance solutions can only be applied to training set
# https://www.kaggle.com/residentmario/undersampling-and-oversampling-imbalanced-data

# Install the library imblearn on Anaconda
# conda install -c conda-forge imbalanced-learn

from collections import Counter
from imblearn.over_sampling import RandomOverSampler
# http://glemaitre.github.io/imbalanced-learn/generated/imblearn.over_sampling.RandomOverSampler.html
from imblearn.over_sampling import SMOTE
# https://imbalanced-learn.org/stable/references/generated/imblearn.over_sampling.SMOTE.html
from imblearn.under_sampling import RandomUnderSampler
# http://glemaitre.github.io/imbalanced-learn/generated/imblearn.under_sampling.RandomUnderSampler.html

# Below is example about how to apply over-/under-sampling on the [whole data set]

# get features and labels
x=df_num.drop('GradeLetter',axis=1)
y=df_num['GradeLetter']

ros = RandomOverSampler(random_state=10)
ros.fit(x, y)
print('\nOriginal dataset shape {}'.format(Counter(y)))
x_resampled, y_resampled = ros.fit_resample(x, y)
print('After oversampling dataset shape {}'.format(Counter(y_resampled)))

print('\nOriginal dataset shape {}'.format(Counter(y)))
ros = RandomUnderSampler(random_state=30)
ros.fit(x, y)
x_resampled, y_resampled = ros.fit_resample(x, y)
print('After undersampling dataset shape {}'.format(Counter(y_resampled)))

# get features and labels, SMOTE can only be applied on numerical feature

ros = SMOTE(k_neighbors=2)
ros.fit(x, y)
print('\nOriginal dataset shape {}'.format(Counter(y)))
x_resampled, y_resampled = ros.fit_resample(x, y)
print('After oversampling by SMOTE dataset shape {}'.format(Counter(y_resampled)))

# In real-world practice, you need to apply these solutions on [training set only]

# Note that, imbalance solutions are only applied on training set.
# In terms of N-folds, you have to split data into train-test splits, and apply the solution
from sklearn.model_selection import KFold
# N-fold data split
# API: https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html

# Assume last column is your label, other columns are features
from sklearn import neighbors
from sklearn.metrics import accuracy_score
import numpy as np

df_knn = df_num.copy(deep=True)
# normalize featureS
for col in df_knn.columns:
    if col != 'GradeLetter':
        # exclude GradeLetter, since it is label in our data
        df_knn[col]=(df_knn[col]-df_knn[col].min())/(df_knn[col].max()-df_knn[col].min())
        
X = df_knn.loc[:, df_knn.columns!='GradeLetter']
y = df_knn.loc[:,'GradeLetter']
# print(X.columns)
# print(type(X))
# print(type(y))

# a new way for N-fold cross validation
kf = KFold(n_splits=5, shuffle=True)
data_5folds = []
for train_index, test_index in kf.split(X,y):
    print("\nTRAIN:", train_index, "TEST:", test_index)
    # get actual data by index
    x_train, x_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    # save data into fold
    fold = [x_train, x_test, y_train, y_test]
    # add each fold data into 5folds
    data_5folds.append(fold)


for k in range(5, 6, 2): 
    acc_5folds = []
    for x_train, x_test, y_train, y_test in data_5folds:
        print('\nOriginal dataset shape {}'.format(Counter(y_train)))
        ros = RandomOverSampler(random_state=10)
        ros.fit(x_train, y_train)
        x_resampled, y_resampled = ros.fit_resample(x_train, y_train)
        print('After oversampling dataset shape {}'.format(Counter(y_resampled)))
        clf=neighbors.KNeighborsClassifier(k, weights='uniform')
        clf.fit(x_resampled, y_resampled)
        y_pred = clf.predict(x_test)
        acc = accuracy_score(y_test, y_pred)
        acc_5folds.append(acc)
    print('k = ',k,'Accuracy on 5-folds: ', np.mean(acc_5folds))
```

```python
# 02. Feature selection: examples of filter method
import matplotlib.pyplot as plt
import seaborn as sns

# For the purpose of filter model, you need to select a metric to tell the quality of the features

# Example 1: when the target variable is a numerical variable
# target variable: Exam
# by using Pearson correlation as selecting criterion

# calculate correlation and show in heatmap
plt.figure(figsize=(12,10))
cor = df_num.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()

#Correlation with output variable
cor_target = abs(cor["Exam"])
#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.5]
print('\nSelected features by Pearson correlation:\n',relevant_features)

# Example 2: when the target variable is a nominal variable
# target variable: GradeLetter
# all other variabels are numerical ones, we can use anova
# if there are nominal features, chi-square can be used

# note that the original nominal variables were converted to binary variables in df_num
# you should test correlation/dependency before the conversion, e.g., chi-square test between nominal variables
# here, for simplicity, I just consider features in df_num are numerical ones, and use ANOVA only

from scipy.stats import f_oneway

X = df_num.drop('GradeLetter', axis=1)
print('\nSelected features by ANOVA:')
for col in X.columns:
    group1 = df_num[df_num['GradeLetter']==0][col]
    group2 = df_num[df_num['GradeLetter']==1][col]
    group3 = df_num[df_num['GradeLetter']==2][col]
    group4 = df_num[df_num['GradeLetter']==3][col]
    stat, p = f_oneway(group1, group2, group3, group4)
    if p < 0.01:
        print(col)
        
# another method to use ANOVA

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

y = df_num['GradeLetter']
x = df_num.drop('GradeLetter', 1)

selector = SelectKBest(f_classif, k=5).fit(x, y)
print('\nSelected features by ANOVA:', selector.get_feature_names_out())
        
# Example 3: when the target variable is a nominal variable
# target variable: GradeLetter
# using feature importance (such as information gain, gini index, etc.) to select features

from sklearn.ensemble import ExtraTreesClassifier
# API, https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.ExtraTreesClassifier.html

y = df_num['GradeLetter']
x = df_num.drop('GradeLetter', 1)

model = ExtraTreesClassifier(criterion='gini')
model.fit(x, y)

values=model.feature_importances_.tolist()
keys=x.columns.tolist()
d = dict(zip(keys, values))
# sort pairs by values descending
s = [(k, d[k]) for k in sorted(d, key=d.get, reverse=True)]


print('\nRanked features by gini importance:\n')
for k, v in s:
    print(k,'\t',v)
    
# many many more methods
# https://scikit-learn.org/stable/modules/classes.html#module-sklearn.feature_selection
# https://scikit-learn.org/stable/modules/feature_selection.html#univariate-feature-selection
```

```python
# 03. Feature selection: examples of wrapper method

# In Wrapper method, you need to use a data science algorithm to evaluate the quality of the features
# This example shows that we can use random forest to select important features

# install library mlxtend
# conda install -c conda-forge mlxtend

from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SequentialFeatureSelector
# search method (forward, backward)
# API, https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.SequentialFeatureSelector.html

feature_selector = SequentialFeatureSelector(RandomForestClassifier(n_jobs=-1),
           n_features_to_select='auto',
           direction='forward',
           scoring='accuracy',
           cv=3)

y = df_num['GradeLetter']
x = df_num.drop('GradeLetter', 1)

feature_selector.fit(x, y)
selected_features= feature_selector.get_support()
print(x.columns)
print(selected_features)
```

```python
# 04. Feature reduction

# Example of PCA

from sklearn.decomposition import PCA


x = df_num.drop('GradeLetter', 1)
y = df_num['GradeLetter']

# feature extraction
# API, https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html
pca = PCA(n_components=10)
fit = pca.fit(x)

# summarize components
# print("Explained Variance: %s") % fit.explained_variance_ratio_
print('\nExplained variance: ', fit.explained_variance_ratio_)

var_sums = []
var_sums.append(0)

for var in fit.explained_variance_ratio_:
    l = len(var_sums)
    v_sum = var_sums[l-1] + var
    var_sums.append(v_sum)

var_sums.remove(0)
print('\nAccmulated variance: ', var_sums)   
print('\nPCAs:\n', fit.components_) 

# select PCA and output new features
# for example, we choose the top-3 PCAs

PCAs = pca.fit_transform(x)
PCAs_selected = PCAs[:,:3]
df_PCAs = pd.DataFrame(data=PCAs_selected, columns=['PC1','PC2','PC3'])
df_PCAs['GraderLetter']=y

display(HTML(df_PCAs.head(10).to_html()))

# write new data to external files
df_PCAs.to_csv('Data_Students_PCA.csv', sep=',')
```
